```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def show_boxes(img, results, color=(0,255,255), thickness=2):
  image = img.copy()
  for result in results:
    cv2.polylines(image, [np.array(result[0], np.int32)], False, color, thickness)
  cv2_imshow(image)
```

```python
show_boxes(img,results)
```

```python
def extract_images(image, results):
  list_of_images = []
  for result in results:
    p0, p1, p2, p3 = result[0]
    x1 = int(p0[0])
    y1 = int(p0[1])
    x2 = int(p2[0])
    y2 = int(p2[1])
    list_of_images.append(image[y1:y2,x1:x2])
  return list_of_images
```

```python
list_of_images = extract_images(img, results)
```

# Загрузка модели

```python
# Импорт библиотек
import requests
from urllib.parse import urlencode
import re
import os
from urllib.parse import unquote
```

```python
# Функция загрузки файла по ссылке на Yandex Disk
def yandex_download(your_link, sub_dir = ''):
  base_url = 'https://cloud-api.yandex.net/v1/disk/public/resources/download?'
  public_key = your_link
  # Получаем загрузочную ссылку
  final_url = base_url + urlencode(dict(public_key=public_key))
  response = requests.get(final_url)
  # Извлекаем имя файла
  filename = re.findall(r'&filename=(\w*.\w*)&', unquote(response.text))[0]
  download_url = response.json()['href']
  # Загружаем файл и сохраняем его
  download_response = requests.get(download_url)
  # Если передан подкаталог
  if len(sub_dir) != 0:
    # Проверяем его существование и создаем если отсутствует
    if not os.path.isdir(sub_dir):
      os.mkdir(sub_dir)
  with open(os.path.join(sub_dir,filename), 'wb') as f:   # Здесь укажите нужный путь к файлу
    f.write(download_response.content)
  # Выводим отчет
  print(f'Файл {filename} загружен в папку /content/{sub_dir}')
  return filename
```
